Recall the definition of the derivative:

Let $I$ be an interval, $f : I \rightarrow \mathbb{R}$, and $c\in I$. If $C$ is the collection of points at which f is differrentiable, then the derivative of $f$ is the function $f' : C \rightarrow \mathbb{R}$ where $$f'(c) = \lim \limits_{x\rightarrow c} \frac{f(x) - f(c)}{x-c},$$ (i.e. Slope = Rise over run).

Alternative definition:

$$f'(x) = \lim \limits_{h\rightarrow 0} \frac{f(x+h) - f(x)}{h}.$$

Let's try something similar:

$f(x) = x^2, \text{ } \forall x\in \mathbb{R}$.

$\begin{align}
f'(c) &= \lim \limits_{x\rightarrow c} \frac{f(x) - f(c)}{x-c} \\
&=  \lim \limits_{x\rightarrow c} \frac{x^2 - c^2}{x-c} \\
&= \lim \limits_{x\rightarrow c} \frac{(x+c)(x-c)}{x-c} \\
&= \lim \limits_{x\rightarrow c} {(x+c)} \\
&= 2c.
\end{align}$
